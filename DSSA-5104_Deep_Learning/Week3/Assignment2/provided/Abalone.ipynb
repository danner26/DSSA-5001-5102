```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
#style.use("ggplot")

def SSR(x,y,a,b):
    #weight_model = a * x + b ## This would be a linear model
    weight_model = a * x ** b
    ssr = np.sum((weight_model-y)**2)
    return ssr

def model(x,y,a,b):
    #weight_model = a * x + b ## This would be a linear model
    weight_model = a * x ** b
    return weight_model

filename = "abalone.data.txt"
# x and y coordinates, whitespace-separated
X = np.loadtxt( filename, delimiter=',', usecols=(2,4) )
diameter = X[:,0] 
weight_observed = X[:,1] 

a,b=5.0,5.0

loss = SSR(diameter,weight_observed,a,b)
weight_model = model(diameter,weight_observed,a,b)    
    
label1=r'Model: W=' + str(a) + r'D$^{' + str(b) + r'}$'
label2=r'SSR = ' + str(loss)
print(label1)
print(label2)

plt.scatter(diameter,weight_observed, marker = 'o', s=150, alpha = 0.5,label='Observed')
plt.scatter(diameter,weight_model,marker='.',color='red',label=label1)
plt.text(0.05,2.2,label1)
plt.text(0.05,2.0,label2)
plt.xlabel('D')
plt.ylabel('W')
plt.legend()

#plt.savefig("abalone.png")
plt.show()
```
